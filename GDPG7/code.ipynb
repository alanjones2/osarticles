```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
# GDP dropped like a stone in Q1 2020

# stats for g7 from OECD: Quarterly National Accounts  : G7 GDP for EI copy
# B1_GE: Gross domestic product - expenditure approach
# The present publication presents time series which extend beyond the date of the United Kingdom’s 
# withdrawal from the European Union on 1 February 2020. In order to maintain consistency over time, 
# the “European Union” aggregate presented here excludes the UK for the entire time series. 
# Interested readers may refer to the Eurostat website for further information on Eurostat’s plans for 
# disseminating EU aggregates and to the Eurostat Database for the actual series.

# https://stats.oecd.org/Index.aspx?QueryId=77241

# split to GPSA, GYSA and VIXOBSA

# GPSA: Growth rate based on seasonally adjusted volume data, percentage change from previous quarter
gpsa = pd.read_csv('GPSA.csv', index_col='Country')    

# GYSA: Growth rate based on seasonally adjusted volume data,  percentage change from same quarter of previous year
gysa = pd.read_csv('GYSA.csv', index_col='Country')  

# VIXOBSA: Volume index, OECD reference year, seasonally adjusted       
vixobsa = pd.read_csv('VIXOBSA.csv', index_col='Country')
```

```python
gpsa
gpsa.T.plot(figsize=(15,5))
```

```python
gysa
gysa.T.plot(figsize=(15,5))
```

```python
# indexed at 2015 - so 2015 is 100

vixobsa.T.plot(figsize=(15,5))
vixobsa
```

```python
# track over 3 years
vixobsa.T[-12:].plot(figsize=(15,5), grid=True)
```

```python
# immediately pre pandemic to end 2020
(vixobsa['Q4-2021']-vixobsa['Q4-2019']).plot.bar()
```

```python
# from worst of pandemic to end 2020
(vixobsa['Q4-2021']-vixobsa['Q2-2020']).plot.bar()
plt.xticks(rotation=45);
```

```python
(vixobsa['Q2-2020']-vixobsa['Q4-2019']).plot.bar(xlabel="") 
plt.grid(axis='y')
plt.suptitle('Negative effect of pandemic',fontsize=16,fontweight='bold',ha='left',x=0.05,y=1.1)
title="""The difference between GDP in Q4 2019 and Q2 2020"""
plt.title(title,ha='left',x=-0.1,y=1)

plt.figtext(1,-0.2,"Data source: OECD, Graphic: AJ", ha='right', fontstyle='italic',fontsize=8)

matplotlib.rcParams['figure.facecolor'] = 'white'

plt.xticks(rotation=45);
```
